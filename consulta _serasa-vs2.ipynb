```python
import pyautogui
import time
import openpyxl

#Abrir navegador
pyautogui.press('win')
time.sleep(1)
pyautogui.write('experian-product-menu')
time.sleep(1)
pyautogui.press('Enter')
time.sleep(4)

#Pessoas
x, y, largura, altura = pyautogui.locateOnScreen('Pessoas.png', grayscale= True, confidence= 0.8)
time.sleep(1)
pyautogui.click(x + largura/2, y + altura/2)
time.sleep(2)

#Concentre
pyautogui.press('tab', presses=12)
time.sleep(1)
pyautogui.press('enter')
time.sleep(4)

#Pesquisa
pyautogui.press('tab', presses=9)
time.sleep(1)
pyautogui.press('enter')
time.sleep(4)

#Importanto a planilha.
workbook = openpyxl.load_workbook('Consulta_Serasa.xlsx')
#Indicando qual aba eu quero utilizar.
pagina_clientes = workbook['Plan1']
#Informando qual a primeira linha de leitura
for linha in pagina_clientes.iter_rows(min_row=2):
    Nome = linha [0].value
    CadastroPF = linha[1].value
    
    #Consulta CPF
    pyautogui.write(str(CadastroPF))
    time.sleep(3)

    #Gerar Relatório
    x, y, largura, altura = pyautogui.locateOnScreen('Gerar_Relatorio.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(5)

    #Salvar ou Imprimir
    x, y, largura, altura = pyautogui.locateOnScreen('Salvar_Imprimir.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(5)
    pyautogui.press('enter')
    time.sleep(2)

    #Tipo de ocorrência
    pyautogui.write(str(Nome))
    time.sleep(3)
    pyautogui.press('enter')
    time.sleep(3)

    #Nova Consulta
    x, y, largura, altura = pyautogui.locateOnScreen('Nova_Consulta.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(5)
    pyautogui.press('tab', presses=3)
    time.sleep(2)
```
